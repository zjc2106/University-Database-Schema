````python
%%sql

use zjc2106_s22_midterm;

DROP TABLE students;
DROP TABLE employees;
DROP VIEW people;
DROP FUNCTION generate_uni;

DROP PROCEDURE create_student;
DROP PROCEDURE create_employee;
DROP PROCEDURE update_student;
DROP PROCEDURE update_employee;


CREATE TABLE students
(
    first_name      varchar(50) NOT NULL,
    middle_name     varchar(50) NULL,
    last_name       varchar(50) NOT NULL,
    email           varchar(50) NOT NULL,
    employee_type   varchar(15) NULL,
    enrollment_year int  CHECK(enrollment_year>2015 AND enrollment_year<2023) NOT NULL,
    uni             varchar(7)  NOT NULL,
    guid            varbinary(36) NOT NULL
    
);


CREATE TABLE employees
(
    first_name      varchar(50) NOT NULL,
    middle_name     varchar(50) NULL,
    last_name       varchar(50) NOT NULL,
    email           varchar(50) NOT NULL,
    employee_type   varchar(15) CHECK(employee_type in ('Professor', 'Lecturer', 'Staff')) NOT NULL,
    enrollment_year int  NULL,
    uni             varchar(7)  NOT NULL,
    guid            varbinary(36) NOT NULL
    
);



CREATE VIEW people (first_name, middle_name, last_name, email, 
                    enrollment_year, employee_type, uni, guid) AS
    SELECT first_name, middle_name, last_name, email, 
    enrollment_year, null, uni, guid
    FROM students
    UNION all
    SELECT first_name, middle_name, last_name, email, 
    null, employee_type, uni, guid
    FROM employees;
    
SET GLOBAL log_bin_trust_function_creators = 1;
    
CREATE FUNCTION generate_uni(first_name varchar(50),
                             middle_name varchar(50),
                             last_name varchar(50))
    RETURNS varchar(7)
    BEGIN
    DECLARE start_uni varchar(3);
    DECLARE count int;
    DECLARE temp_uni varchar(7);
    IF middle_name IS NULL THEN
        SET start_uni = CONCAT(SUBSTRING(first_name, 1, 1), SUBSTRING(last_name, 1, 1));
    ELSE 
        SET start_uni = CONCAT(SUBSTRING(first_name, 1, 1), SUBSTRING(middle_name, 1, 1),
                               SUBSTRING(last_name, 1, 1));
    END IF;
    SET count = 1;
    SET temp_uni = CONCAT(start_uni,count);
    WHILE (SELECT EXISTS(SELECT * from people where uni=temp_uni)) DO
        SET count = count + 1;
        SET temp_uni = CONCAT(start_uni, count);
    END WHILE;
    return temp_uni;
END;

    
CREATE PROCEDURE create_student(IN first_namev varchar(50),
                                IN middle_namev varchar(50),
                                IN last_namev varchar(50),
                                IN emailv varchar(50),
                                IN enrollment_yearv int,
                                OUT uni varchar(7),
                                OUT guid varbinary(36))
BEGIN 
    DECLARE temp_uni varchar(7);
    SET temp_uni = generate_uni(first_namev, middle_namev, last_namev);
    INSERT INTO students (first_name, middle_name, last_name, 
                          email, enrollment_year)
    VALUES (first_namev, middle_namev, last_namev, emailv, enrollment_yearv);
    SELECT guid INTO guid from students where uni = temp_uni;
    SET uni = temp_uni;
END;

CREATE PROCEDURE create_employee(IN first_namev varchar(50),
                                IN middle_namev varchar(50),
                                IN last_namev varchar(50),
                                IN emailv varchar(50),
                                IN employee_typev varchar(15),
                                OUT uni varchar(7),
                                OUT guid varbinary(36))
BEGIN 
    DECLARE temp_uni varchar(7);
    SET temp_uni = generate_uni(first_namev, middle_namev, last_namev);
    INSERT INTO employees (first_name, middle_name, last_name, 
                          email, employee_type)
    VALUES (first_namev, middle_namev, last_namev, emailv, employee_typev);
    SELECT guid INTO guid from students where uni = temp_uni;
    SET uni = temp_uni;
END;

CREATE PROCEDURE update_student(IN univ varchar(7),
                                IN first_namev varchar(50),
                                IN middle_namev varchar(50),
                                IN last_namev varchar(50),
                                IN emailv varchar(50),
                                IN enrollment_yearv int)
BEGIN
    UPDATE students
    SET first_name = IF(first_namev IS NOT NULL, first_namev, first_name), 
        middle_name = IF(middle_namev IS NOT NULL, middle_namev, middle_name), 
        last_name = IF(last_namev IS NOT NULL, last_namev, last_name), 
        email = IF(emailv IS NOT NULL, emailv, email),
        enrollment_year = IF(enrollment_yearv IS NOT NULL, enrollment_yearv, enrollment_year)
    WHERE uni = univ;
END;

CREATE PROCEDURE update_employee(IN univ varchar(7),
                                IN first_namev varchar(50),
                                IN middle_namev varchar(50),
                                IN last_namev varchar(50),
                                IN emailv varchar(50),
                                IN employee_typev varchar(15))
BEGIN
    UPDATE employees
    SET first_name = IF(first_namev IS NOT NULL, first_namev, first_name), 
        middle_name = IF(middle_namev IS NOT NULL, middle_namev, middle_name), 
        last_name = IF(last_namev IS NOT NULL, last_namev, last_name), 
        email = IF(emailv IS NOT NULL, emailv, email),
        employee_type = IF(employee_typev IS NOT NULL, employee_typev, employee_type)
    WHERE uni = univ;
END;

````

````python
%%sql


CREATE TRIGGER before_student_insert BEFORE INSERT
    ON students
    FOR EACH ROW
    BEGIN  
    IF NEW.email IN (SELECT email from people) THEN
        SIGNAL SQLSTATE '50000'
            SET MESSAGE_TEXT = 'insert: email is not unique', MYSQL_ERRNO = 1001;   
    ELSE
        SET new.guid = uuid();
        SET new.uni = generate_uni(new.first_name, new.middle_name, new.last_name);  
    END IF;
END;

CREATE TRIGGER before_student_update BEFORE UPDATE
    ON students
    FOR EACH ROW
    BEGIN  
    IF NEW.email!=old.email AND NEW.email IN (SELECT email from people) THEN
        SIGNAL SQLSTATE '50000'
            SET MESSAGE_TEXT = 'update: email is not unique', MYSQL_ERRNO = 1001;   
    ELSE
        IF new.uni != old.uni OR new.guid !=old.guid THEN
            SIGNAL SQLSTATE '50000'
            SET MESSAGE_TEXT = 'uni and guid are immutable', MYSQL_ERRNO = 1001; 
        END IF;
    END IF;
END;


CREATE TRIGGER before_emp_insert BEFORE INSERT
    ON employees
    FOR EACH ROW
    BEGIN  
    IF NEW.email IN (SELECT email from people)  THEN
        SIGNAL SQLSTATE '50000'
            SET MESSAGE_TEXT = 'bro what email is not unique', MYSQL_ERRNO = 1001;   
    ELSE
        SET new.guid = uuid();
        SET new.uni = generate_uni(new.first_name, new.middle_name, new.last_name);  
    END IF;
END;

CREATE TRIGGER before_emp_update BEFORE UPDATE
    ON employees
    FOR EACH ROW
    BEGIN  
    IF NEW.email!=old.email AND NEW.email IN (SELECT email from people) THEN
        SIGNAL SQLSTATE '50000'
            SET MESSAGE_TEXT = 'update: email is not unique', MYSQL_ERRNO = 1001;   
    ELSE
        IF new.uni != old.uni OR new.guid !=old.guid THEN
            SIGNAL SQLSTATE '50000'
            SET MESSAGE_TEXT = 'uni and guid are immutable', MYSQL_ERRNO = 1001; 
        END IF;
    END IF;
END;
````
